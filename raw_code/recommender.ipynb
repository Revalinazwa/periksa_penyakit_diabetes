## ðŸ“Š EDA (Exploratory Data Analysis)

import dataset

```python
import pandas as pd
# Load dataset
df = pd.read_csv('../dataset/food_data.csv', na_values=['-'])
df.head()
```

convert data type

```python
# Ganti koma jadi titik di seluruh dataframe (hanya kolom string)
df = df.applymap(lambda x: x.replace(',', '.') if isinstance(x, str) else x)

# Daftar kolom numerik
numeric_cols = [
    'Air (g)', 'Energi (Kal)', 'Protein (g)', 'Lemak (g)', 'Karbohidrat (g)', 'Serat (g)', 'Abu (g)',
    'Kalsium (Ca) (mg)', 'Fosfor (P) (mg)', 'Besi (Fe) (mg)', 'Natrium (Na) (mg)', 'Kalium (Ka) (mg)',
    'Tembaga (Cu) (mg)', 'Seng (Zn) (mg)', 'Retinol (vit. A) (mcg)', 'Î²-karoten (mcg)', 
    'Karoten total (mcg)', 'Thiamin (vit. B1) (mg)', 'Riboflavin (vit. B2) (mg)', 'Niasin (mg)', 
    'Vitamin C (mg)', 'BDD (%)'
]

# Convert semua kolom numerik ke float
for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# ceh type data
print(df.dtypes)

```

info dataset

```python
# Cek info awal
print("dataset info: \n")
df.info()

print("\ndataset describe: \n",df.describe())

print("\ndataset shape:",df.shape)
```

cek missing value

```python
# Cek missing values
print(df.isnull().sum())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Distribusi kolom utama
features = ['Energi (Kal)', 'Protein (g)', 'Lemak (g)', 'Karbohidrat (g)', 'Serat (g)', 'Natrium (Na) (mg)']

for feature in features:
    plt.figure(figsize=(8,4))
    sns.histplot(df[feature].dropna(), kde=True)
    plt.title(f'Distribusi {feature}')
    plt.show()

# Korelasi antar nutrisi
plt.figure(figsize=(10,8))
sns.heatmap(df[features].corr(), annot=True, cmap='coolwarm')
plt.title('Korelasi Antar Nutrisi')
plt.show()
```

ðŸ”§ Data Preparation

```python
from sklearn.preprocessing import MinMaxScaler

# Drop rows yang ada NaN pada fitur penting
df_clean = df.dropna(subset=features).reset_index(drop=True)
print("Jumlah data setelah bersih:", df_clean.shape)
df_clean.to_csv('../dataset/food_data_clean.csv')
# Scaling fitur
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(df_clean[features])

# Save scaler
import pickle
with open('../model/scaler.pkl', 'wb') as f:
    pickle.dump(scaler, f)
```

ðŸ”¥ Build Model (KMeans Clustering)

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

wcss = []
silhouette_scores = []
K = range(2, 11)  # K dari 2 sampai 10

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)  # X = data kamu yang sudah discale/diolah
    wcss.append(kmeans.inertia_)  # inertia = WCSS
    score = silhouette_score(X_scaled, kmeans.labels_)
    silhouette_scores.append(score)

# Plot Elbow
plt.figure(figsize=(10,5))
plt.plot(K, wcss, 'o-')
plt.xlabel('Jumlah Cluster (k)')
plt.ylabel('WCSS')
plt.title('Elbow Method Untuk Menentukan k')
plt.show()

# Plot Silhouette Score
plt.figure(figsize=(10,5))
plt.plot(K, silhouette_scores, 'o-')
plt.xlabel('Jumlah Cluster (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score Untuk Menentukan k')
plt.show()
```

ðŸ›  Final Training + Save Model

```python
# Misal hasil elbow method optimal di k=6
final_kmeans = KMeans(n_clusters=3, random_state=42)
final_kmeans.fit(X_scaled)

# Simpan KMeans model
with open('../model/kmeans.pkl', 'wb') as f:
    pickle.dump(final_kmeans, f)

print("Scaler dan Model berhasil disimpan!")
```

ðŸ“ˆ Model Evaluation (Clustering Quality)

```python
from sklearn.metrics import silhouette_score

# Silhouette Score
score = silhouette_score(X_scaled, final_kmeans.labels_)
print(f"Silhouette Score: {score:.4f}")
```

Visualisasi Clustering (PCA 2D)

```python
from sklearn.decomposition import PCA

# PCA 2D
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

plt.figure(figsize=(10,7))
sns.scatterplot(x=X_pca[:,0], y=X_pca[:,1], hue=final_kmeans.labels_, palette='tab10')
plt.title('Visualisasi Cluster Makanan (PCA)')
plt.show()
```
